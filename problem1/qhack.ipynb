```python
import pandas as pd
import numpy as np
```

```python
stocks = pd.read_csv('task-1-stocks.csv')
stocks
```

```python
data = stocks.to_numpy()
multipliers = data[1:] / data[:-1]

# Средние множители для каждой акции
mean = np.mean(multipliers, 0)
# Матрица ковариаций множителей
cov = np.cov(multipliers.T)
```

```python
from scipy import sparse
import json

def calc():
  # Гиперпараметры функции потерь
  q = 1000
  rho = 0.00001

  # Матрица (J в модели Изинга)
  Q = q * cov + rho * (data[0].reshape((-1, 1)) * data[0].reshape((1, -1)))
  # Вектор (h в модели Изинга)
  h = -np.diag(mean) - 2 * rho * 1000000 * np.diag(data[0])

  # Сколько бит используем для бинаризации переменных
  k = 5


  # Вспомогательные матрицы для бинаризации
  a = np.array([2**i for i in range(k)])
  a = a.reshape((-1, 1)) * a.reshape((1, -1))
  # a = a / (4**k)

  b = np.diag([2**i for i in range(k)])
  # b = b / (2**k)

  # Итоговая матрица после бинаризации
  matrix = np.kron(a, Q) + np.kron(b, h)
  matrix = (matrix - np.triu(matrix)).T + np.triu(matrix)
  # print(matrix)

  m = sparse.coo_matrix(matrix)

  # Сохраняем как edgelist
  with open("matrix.sparse", "w") as f:
    f.write(f"{matrix.shape[0]} {m.getnnz()}")
    for i in range(m.getnnz()):
      f.write(f"\n{m.row[i] + 1} {m.col[i] + 1} {m.data[i]}")


  # Запуск в облаке
  !python3 run_qiopt_cli.py -i matrix.sparse -u user_data.txt -o output.txt


  # Сохраняем результат
  with open('output.txt', 'r') as f:
    result = json.load(f)


  # Приводим результат к читаемому виду
  solution = np.array(result["Solution"])
  solution = solution.reshape((k, -1))

  pows = np.array([2**i for i in range(k)]).reshape((-1, 1))

  # Получившийся вектор
  vector = np.sum((solution * pows), 0)
  # vector = np.sum((solution * pows), 0) / 2**k


  print(result["Objective"])
  # vector = np.round(vector / np.sum((data[0] * vector)) * 1000000)
  return vector
```

```python
vector = calc()
```
